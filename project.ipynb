**PART 1:MACHINE LEARNING MODEL**

  **Flight Ticket Price Prediction**


1. Problem Type

 Regression Problem ‚Üí Predict Price

2. Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

**import** **dataset**



```python
from google.colab import files
uploaded = files.upload()
```

3. Load Dataset

```python
df = pd.read_csv("airline_data.csv")
print(df.columns)
```

4. Features and Target

```python
X = df.drop("Price", axis=1)
y = df["Price"]
```

5. Preprocessing

```python
num_features = ["Duration"]
cat_features = ["Airline", "Source", "Destination", "Total_Stops"]


preprocessor = ColumnTransformer([
("num", StandardScaler(), num_features),
("cat", OneHotEncoder(handle_unknown="ignore"), cat_features)
])
```

6. Model Pipeline

```python
model = Pipeline([
("prep", preprocessor),
("rf", RandomForestRegressor(n_estimators=200))
])
```

7. Train Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model.fit(X_train, y_train)
```

8. Evaluate

```python
preds = model.predict(X_test)
print("MAE:", mean_absolute_error(y_test, preds))
```

**PART 5: FLIGHT RECOMMENDATION SYSTEM**

Logic

Recommend cheapest or fastest flights based on user input

```python
def recommend_flights(df, source, destination):
  results = df[(df["Source"] == source) & (df["Destination"] == destination)]
  return results.sort_values("Price").head(5)


recommend_flights(df, "Delhi", "Mumbai")
```
